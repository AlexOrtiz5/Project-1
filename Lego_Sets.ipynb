```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
grouped_data = Clean_Lego_df.groupby('year')['set_price'].mean().reset_index()
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='year', y='set_price',data=grouped_data, marker='o', color='b')
plt.title('Average Lego Set Prices Over the Years')
plt.xlabel('year')
plt.ylabel('Average Price')
plt.show()
```
