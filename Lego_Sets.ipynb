### 2. Popular Themes and Ratings: Check to see if any themes have a tendency to score better than others or to accumulate more reviews. Make an educated guess as to whether popular themes sustain high ratings year after year or whether there is variation.

```python
# Group by the Theme and the mean of the Star Rating
Theme_rating = Clean_Lego_df.groupby('theme_name')['star_rating'].mean().round(2)
# Group by the Theme and sum of the Number of reviews
Theme_reviews = Clean_Lego_df.groupby('theme_name')['number_of_reviews'].mean().round(2)
# Merge the Rating and Reviews
Theme_merged = pd.merge(Theme_rating, Theme_reviews, on='theme_name', how='inner')
# Sort the values
Theme_sorted = Theme_merged.sort_values(by='number_of_reviews', ascending=False).head(10)
Theme_sorted
```

```python
# Plotting top 10 themes by total reviews and their average star ratings
Theme_sorted.plot(kind='bar', y=['star_rating', 'number_of_reviews'])
plt.title('Top 10 Themes by Reviews and Star Ratings')
plt.xlabel('Theme')
plt.ylabel('Values')
plt.legend(['Star rating', 'Total Reviews'])
plt.grid(axis='y')
plt.show()
```

```python
# Plotting top 10 themes by total reviews 
Theme_sorted.plot(kind='bar', y= 'number_of_reviews')
plt.title('Top 10 Themes by Reviews')
plt.xlabel('Theme')
plt.ylabel('Values')
plt.grid(axis='y')
plt.show()
```

```python
# Plotting top 10 themes by average star ratings
Theme_sorted.plot(kind='bar', y= 'star_rating')
plt.title('Top 10 Themes by Star Ratings')
plt.xlabel('Theme')
plt.ylabel('Values')
plt.grid(axis='y')
plt.ylim(bottom=3)
plt.show()
```

At the end after this three graphs. The second and third show us the to 10 themes. Where they are different do to the analysis that were done on each column that is being display. For instance the second plot show us the 10 themes in order of the Number of Reviews. While the third plot show us a almost different order do to the Star Ratings of the 10 Themes.

### 3. Impact of Part Material or Color on Set Price: Make a conjecture regarding the potential influence of part material or color on the set price. Sets with specific materials or colors, for example, may have varying prices because of their scarcity or desirability.

Impact of the Part Material on Set Price

```python
# Group by the Part material and calculate the mean
Material_price = Clean_Lego_df.groupby('part_material')['set_price'].mean().round(2)
# Sort the Materials
Material_sorted = Material_price.sort_values(ascending=False)
Material_sorted
```

```python
# Plotting the average set prices based on part material
Material_sorted.plot(kind='bar')
plt.title('Average Set Prices by Part Material')
plt.xlabel('Part Material')
plt.ylabel('Average Set Price')
plt.grid(axis='y')
plt.show()
```

This plot is showing us that the price really is impacted by the material of the lego. Were if the lego is made out of metal it is going to cost more compare to plastic that follows it in price.

Impact of the Part Color on Set Price

```python
# Group by the Color and calculate the mean
Color_price = Clean_Lego_df.groupby('part_color')['set_price'].mean().round(2)
# Sort the Materials
Color_sorted = Color_price.sort_values(ascending=False)
Color_sorted
```

```python
# Plotting the average set prices based on part color
Color_sorted.plot(kind='bar')
plt.title('Average Set Prices by Part Color')
plt.xlabel('Part Color')
plt.ylabel('Average Set Price')
plt.grid(axis='y')
plt.show()
```

This plot is showing that there is a significance of great and lower cost depending on the color that is being use on lego
